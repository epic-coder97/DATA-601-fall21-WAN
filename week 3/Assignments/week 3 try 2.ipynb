```python
import numpy as np
import pandas as pd
from faker.providers.person.en import Provider
from faker import Faker
import random
from datetime import date 
fake = Faker()
```

```python
input_num = int(input('Enter the number of Entries: '))
data =[]
for i in range (input_num):
    #print (i)
    a = fake.name()
    #df = pd.DataFrame(a)
    email_data = ["@gmail.com", "@yahoo.com", "@hotmail.com", "@aol.com", "@outlook.com"]
    #using names to make email address realistic
    e = a.replace(" ", "").lower() + random.choice(email_data)
    addr = fake.address()
    phno = fake.phone_number()
    ###print(phno)
    
    dob = fake.date_of_birth(tzinfo=None, minimum_age=18, maximum_age=100)
    ###print(dob)
    today = date.today()
    #calculating age of the person
    age = today.year - dob.year - ((today.month, today.day) < (dob.month, dob.day))
    ###print(age)
    
    #based on age calculating number of kids
    if age < 18:
        kids = 0
    elif age < 30  and age > 18:
        kids = random.randrange(0,2,1)
    elif age > 30 and age < 40:
        kids = random.randrange(0,4,1)
    else : 
        kids = random.randrange(0,5,1)
    
    ###print("kids :", kids)   
    
    
    #MEDIAN HOUSE HOLD INCOME IS $64,000

    income = random.randrange(63000 , 380000 ,9000)
    ###print("$",income)

    if income > 150000 and age > 25:
        house = "own"
    else:
        house = "rent"
    
    ###print ("house : " , house)   
    #speeding ticket 

    s_ticket = random.randrange(0 , 10 , 1)
    ###print("speeding Ticket :",s_ticket)

    data1 = [a,e,addr,phno,age,kids,income,house,s_ticket]
    data.append(data1)
    df = pd.DataFrame(data, columns=['Name','Email','Address','Phone No','Age','kids','Income','Rent/own','Speeding tickets'])
    
df
    
```

```python
df.to_csv('my_data.csv')
```

```python
import requests
import pandas as pd
import json
from pandas.io.json import json_normalize
```

```python
#getting population data from DATA USA
#https://datausa.io/about/api/
```

```python
pd.show_versions()
```

```python
data_usa_pop = 'https://datausa.io/api/data?measures=Population'
#update the drilldowns to either Nation or State, also can update the Year to either specfic year or "latest"
response = requests.get(data_usa_pop, params = {"drilldowns":"State","Year":"2016"})

api_raw = response.json()
json_str = json.dumps(api_raw)
resp = json.loads(json_str)
data = resp['data']
df = pd.json_normalize(data)
df
```

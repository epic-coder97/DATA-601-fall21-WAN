```python
import openpyxl 
import pandas as pd
```

```python
#week_05_homework_XLSX_openpyxl
wb = openpyxl.load_workbook('week_05_homework_XLSX_openpyxl.xlsx')

sheet_name = []
#storing sheet names in a list so I can later call them using index numbers
sheet_name = wb.sheetnames

#calling sheet names by index
print(sheet_name[1])

#printing all sheets
print("Sheets names in the week_05_homework_XLSX_openpyxl.xlsx : ",sheet_name)
```

## Storing <i>Patient_info</i> in DataFrame 

```python
sheet = wb[sheet_name[0]] # calling PATIENT_INFO

#storing records in dataframe
data1 = sheet.values

# Get the first line in file as a header line
columns = next(data1)[0:]

# Create a DataFrame based on the second and subsequent lines of data
df1 = pd.DataFrame(data1, columns=columns)

#df = pd.DataFrame(sheet.values)
df1
```

## Storing <i>  Records </i> in DataFrame 

```python
sheet = wb[sheet_name[1]] # calling records
#storing records in dataframe
data2 = sheet.values
 # Get the first line in file as a header line
columns = next(data2)[0:]
# Create a DataFrame based on the second and subsequent lines of data
df2 = pd.DataFrame(data2, columns=columns)
#df = pd.DataFrame(sheet.values)
df2
```

Here we will perform a right join as :
<li>For each patient identified by "p_id" listed in the "records" sheet that  <b>don't</b> exist in "patient info," create a new row in "patient info". The row from "records" should be copied to the row in "patient info."</li>
<li>For each patient listed in the "records" sheet where "p_id" <b>appears</b> "patient info" under "patient id," copy the data in the row from records to "patient info" sheet</li>


```python
#Merging the Dataframes on P_id and patient_info 
df_merge = pd.merge(df1, df2, how='right', left_on = 'patient id', right_on = 'p_id')

# droping p_id
df_mergedf = df_merge.drop(['p_id'], axis = 1)


#exporting to new .xlsx file called new.xlsx. Without the the first index column.
df_mergedf.to_excel('new.xlsx', index=False , sheet_name= 'new patient info')
```

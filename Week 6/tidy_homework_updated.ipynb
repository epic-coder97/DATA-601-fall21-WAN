# data 1
https://data.ny.gov/Energy-Environment/Solid-Waste-Management-Facilities/2fni-raj8

```python
df1 = pandas.read_csv('https://data.ny.gov/api/views/2fni-raj8/rows.csv?accessType=DOWNLOAD&sorting=true')
df1.shape
```

```python
df1.info()
```

### Is this data tidy or not?

(Yes or no)

Ans:  <b>NO</b>

*if the data is not tidy, then convert the data to a tidy dataframe*

```python
# droping null values
print(df1.isna().sum())
df1.dropna(how='any', inplace=True)

#phone number is stored as a float value - converting into str
df1['Phone Number'] = df1['Phone Number'].astype(str)
#removing the decimal
df1['Phone Number'] = df1['Phone Number'].str.split('.').str[0]

df1.head()
```

# data 4
https://catalog.data.gov/dataset/new-york-city-leading-causes-of-death-ce97f


```python
df6 = pandas.read_csv('https://data.cityofnewyork.us/api/views/jb7j-dtam/rows.csv?accessType=DOWNLOAD')
df6.shape
```

```python
df6
```

```python
df6.info()
```

### Is this data tidy or not?

(Yes or no)
<b>NO</b>
*if the data is not tidy, then convert the data to a tidy dataframe*

```python
#droping Nan
print(df6.isna().sum())
df6.dropna(how='any', inplace=True)
```

```python
drop = df6[df6.Deaths == '.'].index
df6.drop(drop, inplace = True)
df6
```

```python
df6.info()
```

```python
#droping rows with '.' as values
drop2 = df6[df6['Death Rate'] == '.'].index
df6.drop(drop2, inplace = True)
drop3 = df6[df6['Age Adjusted Death Rate'] == '.'].index
df6.drop(drop3, inplace = True)
```

```python
#Data type of Deaths , Deathrate , Age Adjusted Death rate is not correct
df6['Deaths'] = df6['Deaths'].astype(int)
df6['Death Rate']= df6['Death Rate'].astype(float)
df6['Age Adjusted Death Rate'] = df6['Age Adjusted Death Rate'].astype(float)
df6.info()
```
